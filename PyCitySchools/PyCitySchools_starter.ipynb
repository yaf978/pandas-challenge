```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"

student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)

student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete

```

```python
# The total number of schools
school_count=len(school_data)

# The total number of students
total_number_of_students = len(student_data)

# The total budget
district_total_budget = school_data["budget"].sum()

# The average math score
average_math_score = school_data_complete["math_score"].mean()

# The average reading score
average_reading_score = school_data_complete["reading_score"].mean()

# The percentage of students with a passing math score (70 or greater)
percentage_math70 = len(school_data_complete.loc[(school_data_complete["math_score"]>=70)])/total_number_of_students * 100


# The percentage of students with a passing reading score (70 or greater)
percentage_reading70 = len(school_data_complete.loc[(school_data_complete["reading_score"]>=70)])/total_number_of_students * 100


# The percentage of students who passed math and reading (% Overall Passing)
percentage_pass =  len(school_data_complete.loc[(school_data_complete["reading_score"]>=70)&(school_data_complete["math_score"]>=70)])/total_number_of_students * 100
                      
# Creating a dataframe to hold the above results
all_schools_summary_df = pd.DataFrame({"Total Schools":[school_count],"Total Students":[total_number_of_students],
                                       "Total Budget":[district_total_budget],"Average Math Score":[average_math_score],
                                       "Average Reading Score":[average_reading_score],
                                       "% Passing Math":[percentage_math70],"% Passing Reading":[percentage_reading70],
                                       "% Overall Passing":[percentage_pass]})

all_schools_summary_df
# YOU HAVE TO FORMAT THE DATAFRAME--------------------------------------++++++++++
```

```python

school_grouped = school_data_complete.groupby("school_name")

school_type = school_grouped["type"].unique()
school_budget = school_grouped['budget'].unique()

# the total number of schools #
# the total number of students #
total_students = school_grouped["Student ID"].count()

# the total budget #
total_school_budget =  school_grouped["budget"].mean()

# Per student budget
per_student_budget = total_school_budget/total_students

# the average math score#
avarge_math_score = school_grouped["math_score"].mean()


# the average reading score #
avarge_reading_score = school_grouped["reading_score"].mean()


# Math Pass Percentage # Reading Pass Percentage
school_pass_math = school_data_complete.loc[school_data_complete['math_score'] >= 70] 

school_pass_reading = school_data_complete.loc[school_data_complete['reading_score'] >= 70] 

school_math_grp = school_pass_math.groupby(['school_name']).count()

school_reading_grp = school_pass_reading.groupby(['school_name']).count()

math_percent_pass = (school_math_grp['math_score']/total_students)*100

reading_percent_pass = (school_reading_grp['reading_score']/total_students)*100

# The percentage of students who passed math **and** reading #

school_pass_both = school_pass_reading.loc[school_pass_reading['math_score'] >= 70]

school_pass_both_grp = school_pass_both.groupby(['school_name']).count() 

both_percent_pass = (school_pass_both_grp['reading_score'] /total_students) * 100



# dataframe to hold the above results ##

school_summary = pd.DataFrame({
     "School Type":school_type,
     "Total Students":total_students,
     "Toatal School Budget":school_budget,
     "Per Student Budget": per_student_budget,
     "Average Math Score": avarge_math_score, 
     "Average Reading Score":avarge_reading_score, 
     "% Passing Math":math_percent_pass, 
     "% Passing Reading":reading_percent_pass,
     "% Overall Passing":both_percent_pass})

school_summary.index.name = None



school_summary

# YOU HAVE TO FORMAT THE DATAFRAME--------------------------------------++++++++++
```
